```python
import json
```

```python
comprehensive_data = {}

for i in range(1990, 2023):
    
    yearly_rosters = {}
    
    file_path = f'../data/{i}-{i+1}.json'

    with open(file_path, 'r') as f:
        data = json.load(f)
        
    yearly_rosters['Initial Rosters'] = data
    yearly_rosters['Updated Rosters'] = {}
    yearly_rosters['Final Rosters'] = data

    comprehensive_data[f'{i}-{i+1}'] = yearly_rosters
    
output_file_path = f'data/comprehensive.json'
with open(output_file_path, 'w') as output_file:
    json.dump(comprehensive_data, output_file, indent=2)
```
