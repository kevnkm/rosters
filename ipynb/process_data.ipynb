```python
import json
```

```python
for i in range(1990, 2023):
    

    file_path = f'data/raw/{i}-{i+1}.json'

    with open(file_path, 'r') as f:
        data = json.load(f)

    teams_data = {}

    for player in data:
        team_name = player["Team"]
        if team_name != "None" and team_name:
            if team_name not in teams_data:
                teams_data[team_name] = []
            teams_data[team_name].append(player)

    teams_list = list(teams_data.items())

    output_file_path = f'{i}-{i+1}.json'
    with open(output_file_path, 'w') as output_file:
        json.dump(teams_data, output_file, indent=2)
```
