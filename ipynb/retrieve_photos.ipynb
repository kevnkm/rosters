```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, NoSuchElementException

import requests
```

```python
driver = webdriver.Safari()

url = "https://hoopshype.com/player/stephen-curry/salary/"
driver.get(url)

div_element = driver.execute_script("return document.querySelector('.player-headshot')")
image_element = driver.execute_script("return arguments[0].querySelector('img')", div_element)

image_url = image_element.get_attribute('src')

response = requests.get(image_url)
if response.status_code == 200:
    with open('downloaded_images/stephen-curry.png', 'wb') as f:
        f.write(response.content)

driver.quit()
```

```python
def retrieve_headshot_photo(name):

    url = "https://hoopshype.com/player/" + name + "/salary/"
    driver.get(url)

    try:
        div_element = WebDriverWait(driver, 3).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'player-headshot'))
        )

        image_element = div_element.find_element(By.TAG_NAME, 'img')
        image_url = image_element.get_attribute('src')

        response = requests.get(image_url)
        if response.status_code == 200:
            with open('downloaded_image.jpg', 'wb') as f:
                f.write(response.content)

    except (TimeoutException, NoSuchElementException):
        print(f"Could not find headshot photo for {name}")
```

```python
driver.quit()
```

```python
driver = webdriver.Safari()

retrieve_headshot_photo('lebron-james')

driver.quit()
```
